```python
# Packages
import pandas as pd
import altair as alt
import numpy as np

# Constants
VOTES = "../data/raw/mk8dx_votes.csv"
COURSES = "../data/raw/mk8dx_courses.csv"

# Main
votes = pd.read_csv(VOTES)
print('Votes dataframe:')
print(votes.head())
print('========\n')

courses = pd.read_csv(COURSES)
print('Courses dataframe:')
print(courses.head())
print('========\n')
```

```python
def join_dataframes(df1, df2, df1_join, df2_join):
    merged_df = pd.merge(df1, df2, left_on=df1_join, right_on=df2_join)
    merged_df.drop(columns=[df2_join], inplace=True)
    return merged_df

online_votes = join_dataframes(votes, courses, 'selected_course', 'course')
online_votes.head()
```

```python
def add_boolean_column(df, col1, col2, new_col_name):
    df[new_col_name] = df[col1] == df[col2]
    return df

online_votes = add_boolean_column(online_votes, 'selected_course', 'voted_course', 'selected')
online_votes.head()
```

```python
online_votes.to_csv('../data/mk8dx_online.csv', index=False)
```
